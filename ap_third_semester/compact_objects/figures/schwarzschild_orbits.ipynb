```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def lam(x, Gamma):
    return(x * np.sqrt((x * Gamma + 1)/(x-1)))
```

```python
x = np.linspace(0, 12, num=1000)

Gammas = - 1/9 + np.linspace(-.1, .1, num=10)

for G in Gammas:
    lambda_of_x = lam(x, Gamma=G)
    plt.plot(x, lambda_of_x)
    plt.plot(x, -lambda_of_x)
plt.yscale('symlog')
```
