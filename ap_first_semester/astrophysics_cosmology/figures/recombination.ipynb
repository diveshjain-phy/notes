```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.optimize import fsolve
import scipy.constants as sc
from astropy.constants import codata2018 as ac
from astropy.constants import iau2015 as aa
import astropy.units as u
from astropy.cosmology import Planck15 as cosmo
```

```python
z= np.linspace(1000, 2000)

rhoC = 3 * cosmo.H(z)**2 / (8 * np.pi * ac.G)
n_e = rhoC * cosmo.Ob(z) / ac.m_p

argument = (1/n_e  *(cosmo.Tcmb(z) * ac.k_B * ac.m_e / 2/np.pi/ac.hbar**2)**(3/2) * np.exp(- 13.6 *u.eV / (ac.k_B * cosmo.Tcmb(z)))).to(u.dimensionless_unscaled)

x_ratio = [fsolve(lambda x : x**2 / (1-x) - a, .5) for a in argument]

plt.plot(z, x_ratio)
```
